```python
import yaml
import pandas as pd
import os
```

```python
with open("../../../configs/data/regime_universe.yaml", "r") as f:
    regime_cfg = yaml.safe_load(f)
```

```python
for series_id, series_cfg in regime_cfg["regime_universe"]["series"].items():
    for transform in series_cfg["transforms"]:
        transform_parts = transform.split("_")
        transform_type = transform_parts[0]
        transform_window = transform_parts[1]
        print(f"type: ", transform_type, f"{"window:":>10}", transform_window)
```

```python
df_regime_raw = pd.read_parquet(os.path.join("../../../", regime_cfg["regime_universe"]["output_path"]))
df_regime_raw
```
